```python
import pandas as pd
import numpy as np

data = pd.read_csv("Final_idea_dataset.csv")

data.head()
```

```python
data.info()
```

```python
data_sorted = data.sort_values(by='Total_Score')
```

```python
top_500 = data_sorted.tail(500)  # Bottom 500 (lowest scores)
bottom_500 = data_sorted.head(500)  # Top 500 (highest scores)
```

```python
top_500.head()
```

```python
bottom_500.head()
```

```python
bottom_500.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plotting histograms for 'Total Score' of top 500 and bottom 500
plt.figure(figsize=(12, 6))
sns.histplot(top_500['Total_Score'], kde=True, color='green', label='Top 500')
sns.histplot(bottom_500['Total_Score'], kde=True, color='red', label='Bottom 500')
plt.title('Distribution of Total Score for Top 500 vs Bottom 500')
plt.xlabel('Total_Score')
plt.legend()
plt.show()
```

# In the above graph, it shows how the top 500 and bottom 500 datasets are distributed based on the Total_Score, that they have acheived.

# 

# Boxplot for 'Total_Score' of top 500 and bottom 500

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Total_Score', data=pd.concat([top_500.assign(dataset='Top 500'), bottom_500.assign(dataset='Bottom 500')]), hue='dataset')
plt.title('Boxplot of Total Score for Top 500 vs Bottom 500')
plt.ylabel('Total_Score')
plt.show()
```

# 

# Plotting Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='dataset', y='Total_Score', data=pd.concat([top_500.assign(dataset='Top 500'), bottom_500.assign(dataset='Bottom 500')]))
plt.title('Violin Plot of Total Score for Top 500 vs Bottom 500')
plt.ylabel('Total_Score')
plt.show()
```

# In the above Violin Graph, it can be clearly seen for the Top 500 Data, the Median is close to 0.75, whereas for the Bottom 500, it is close to 0.68 as the data in bottom 500 has different Total_Score which decreases the median value. Also the Black line interprets how the data is distributed, so for the Bottom 500 the minimum value is 0.0 & whereas maximum value is around 0.74.

# 

# Plotting Bar Graph

```python
plt.figure(figsize=(8, 6))
top_500_mean = top_500['Total_Score'].mean()
bottom_500_mean = bottom_500['Total_Score'].mean()
plt.bar(['Top 500', 'Bottom 500'], [top_500_mean, bottom_500_mean], color=['blue', 'red'])
plt.title('Mean Total Score for Top 500 vs Bottom 500')
plt.ylabel('Mean Total Score')
plt.show()
```

# Through this bar graph, we can conclude that Mean value for the Top 500 is more than 0.72 and the same value for the Bottom 500 is 0.65

# 

# Plotting Heatmap

```python
plt.figure(figsize=(8, 6))
data_concat = pd.concat([top_500, bottom_500])
sns.heatmap(data_concat.corr(), annot=True, cmap='Reds')
plt.title('Correlation Heatmap of Top 500 and Bottom 500 Data')
plt.show()
```

# From the above heatmap, we can conclude that Total_Score has the highest correlation between Environmental Impact Score (EIS) ,Financial Viability Score (FVS), Feasibility & Scalability Score (FSS), Innovation & Novelty Score (INS) &  Social Impact Score (SIS) with value 0.97, 0.98, 0.99, 0.97 & 0.96 respectively.

```python
# Concatenating both DataFrames
combined_data = pd.concat([top_500, bottom_500])

print(combined_data.columns)
```

# Plotting Pairplot graph

```python
# Pairplot (Scatter Plot Matrix) for numerical variables
sns.pairplot(combined_data[['EIS', 'FVS', 'FSS', 'INS', 'SIS', 'Total_Score']])
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```

# Pairplot graph allow us to plot pairwise relationship between variables, which in our case is EIS, FVS, FSS, INS, SIS & Total_Score.
